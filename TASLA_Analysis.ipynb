## Task-1- Technical Requirements

# Rolling Volatility Windows

```python
# Rolling volatility (20-period window)
df['Rolling_Volatility'] = df['Close'].rolling(window=5).std()
```

```python
df.head(50)
```

```python
# VWAP calculation
df['VWAP'] = (df['Volume'] * df['Close']).cumsum() / df['Volume'].cumsum()
```

```python
df.head()
```

```python
# 20-period moving average
df['MA20'] = df['Close'].rolling(window=20).mean()
# 50-period moving average
df['MA50'] = df['Close'].rolling(window=50).mean()
```

```python
df.head(50)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], label='Close Price', color='blue')
plt.plot(df['Date'], df['Rolling_Volatility'], label='Rolling Volatility', linestyle='--', color='orange')
plt.plot(df['Date'], df['VWAP'], label='VWAP', linestyle='--', color='green')
plt.plot(df['Date'], df['MA20'], label='20-Period MA', linestyle='--', color='purple')
plt.plot(df['Date'], df['MA50'], label='50-Period MA', linestyle='--', color='red')
plt.legend()
plt.title('Stock Metrics Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Calculate correlation matrix
correlation_matrix = df[['Close', 'Volume', 'VWAP', 'Rolling_Volatility']].corr()
# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Between Metrics')
plt.show()
```

```python
# Standard deviation of the closing price
std_dev = df['Close'].std()
# Skewness of the closing price
skewness = df['Close'].skew()
print(f"Standard Deviation: {std_dev}")
print(f"Skewness: {skewness}")
```

```python
# Scatter plot to visualize the relationship
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df['VWAP'], df['Rolling_Volatility'], alpha=0.6, color='blue')
plt.title('Relationship Between VWAP and Rolling Volatility')
plt.xlabel('VWAP')
plt.ylabel('Rolling Volatility')
plt.show()
```
